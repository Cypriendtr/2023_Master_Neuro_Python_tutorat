```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
array=np.random.choice((0,1),1000,p=[0.75,0.25])
kernel=np.array([0,0.1,0.5,1,1,0.5,0.1,0])
```

```python
def AnalySignal(array=array,kernel=kernel):
    a=[]
    for n in range(len(array)):
        if n<len(kernel):
            a.append(sum(array[0:n+1]*kernel[-(n+1):]))
        elif n>=(len(array)-len(kernel)):
            a.append(sum(array[n:]*kernel[:len(kernel)-((len(kernel)+n)-len(array))]))
        else:
            a.append(sum(array[n:n+len(kernel)]*kernel))

    fig,ax=plt.subplots(figsize=(20,10))
    ax.plot(array,"lightgrey")
    ax.plot(a)
    ax.set_title(f"{kernel =}")
    plt.show()
```

```python
kernelX=np.array([0,0.05,0.1,0.5,0.7,0.9,1,1,0.9,0.7,0.5,0.1,0.05,0])
AnalySignal(kernel=kernelX)
```
