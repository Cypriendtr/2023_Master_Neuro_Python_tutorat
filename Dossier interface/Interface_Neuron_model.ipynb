```python
import tkinter as Tk
from tkinter import *
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (
    FigureCanvasTkAgg, NavigationToolbar2Tk)
from matplotlib.figure import Figure
import numpy as np
from PIL import *
import time as t

root=Tk();root['bg']='#81c6eb'
root.title("Page d'acces à l'application")
root.geometry("310x310")
root.geometry("+400+400")
Code=Label(root,text="Veuiller entrer le code secret \n pour acceder à l'application",bg='#81c6eb',font=('calibiri',20,'bold') )
Code.pack(padx=10,pady=5)
CODE_ENTREE_APP=StringVar()
Code_entree_app=Entry(root,textvariable=CODE_ENTREE_APP)
Code_entree_app.pack(padx=10,pady=50)
indice_code_secret=Label(root,text="Le code est la commande pour faire afficher \n hello world en python",bg='#81c6eb',font=('calibiri',14))
indice_code_secret.place(x=10,y=160)
welldone=PhotoImage(file="welldone.gif")
photo=PhotoImage(file="Dev.png")
def verifmdp():
    if str(Code_entree_app.get())=="""print("hello world")""":
        def demarrer_app():            
            root1=Toplevel(root,bg='#81c6eb')
            root1.title("Page d'accueil")
            root1.geometry("800x780")
            root1.resizable(False, False)
            root1.geometry("+400+400")
            clic=IntVar()
            clic.set(0)
            def time():
                Localtime=t.localtime()
                date_heure=t.strftime("%m/%d/%Y   %H:%M:%S %p",Localtime)
                DATE.config(text=date_heure)
                DATE.after(100,time)
            DATE=Label(root1,font=('calibiri',20,'bold'))
            DATE.pack()
            time()
            felicitation="""Bien ouej \n tu as trouvé \n un easter egg"""
            def bienouej():
                titlecam=Label(root1,text=felicitation)
                titlecam.place(x=670,y=550)  
            
            def show1():
                
                titledev=Label(root1,text="Photo des developpeurs en Cours")
                titledev.place(x=250,y=280)
                clic.set(int(clic.get())+1)
                CLIC=clic.get()
                if int(CLIC) == 10 :
                    BJ=Button(root1,image=welldone,command=bienouej)
                    BJ.place(x=670,y=650)
            

            Bienvenue=Label(root1,text="Bienvenue sur cette interface de modélisation de neurone \n réalisé dans le cadre du Tutorat Code"
                            ,bg='#81c6eb',font=('calibiri',20,'bold'))
            Bienvenue.pack(padx=20, pady=10)
            def Window_model_int_and_fire():
                int_fire_W=Toplevel(root,bg="light blue")
                int_fire_W.title("Modele de Neurone: \n Integrate & Fire")
                int_fire_W.geometry("800x780");int_fire_W.resizable(False, False);int_fire_W.geometry("+400+400")

                RESOLUTION_GRAPH_INT_FIRE=IntVar()
                START_INT_FIRE=IntVar()
                DUREESTIM_INT_FIRE=IntVar()
                I_INPUT_INT_FIRE=IntVar()
                VTHRESHOLD_INT_FIRE=IntVar()
                CONSTANTETEMPS_INT_FIRE=IntVar()
                VREPOS_INT_FIRE=IntVar()
                RESISTANCE_MEMB_INT_FIRE=IntVar()
                VRESET_INT_FIRE=IntVar()
                VSPIKE_INT_FIRE=IntVar()
                DUREE_RECORD_INT_FIRE=IntVar()


                Int_fire_veuillez_choisir_resolution_graph=Label(int_fire_W,text="Resol graph(ms)")
                Int_fire_veuillez_choisir_debut_stim=Label(int_fire_W,text="Début stim (ms)")
                Int_fire_veuillez_choisir_durée_stim=Label(int_fire_W,text="Durée Stim (ms)")
                Int_fire_veuillez_choisir_I_input=Label(int_fire_W,text="Intensité stim(pA)")
                Int_fire_veuillez_choisir_cst_temps=Label(int_fire_W,text="Cst de Temps")  
                Int_fire_veuillez_choisir_VThreshold=Label(int_fire_W,text="Seuil PA")
                Int_fire_veuillez_choisir_Vrepos=Label(int_fire_W,text="V repos")
                Int_fire_veuillez_choisir_R_memb=Label(int_fire_W,text="R Memb")
                Int_fire_veuillez_choisir_Vreset=Label(int_fire_W,text="V reset")
                Int_fire_veuillez_choisir_Vspike=Label(int_fire_W,text="V spike")
                Int_fire_veuillez_choisir_durée_record=Label(int_fire_W,text='Temps record (s)')

                Resolution_graph_int_fire=Entry(int_fire_W,textvariable=RESOLUTION_GRAPH_INT_FIRE,width=6)
                Start_int_fire=Entry(int_fire_W,textvariable=START_INT_FIRE,width=6)
                Durée_stim_int_fire=Entry(int_fire_W,textvariable=DUREESTIM_INT_FIRE,width=6)
                I_input_int_fire=Entry(int_fire_W,textvariable=I_INPUT_INT_FIRE,width=6)
                Vthreshold_int_fire=Entry(int_fire_W,textvariable=VTHRESHOLD_INT_FIRE,width=6)
                Vrepos_int_fire=Entry(int_fire_W,textvariable=VREPOS_INT_FIRE,width=6)
                Resistance_memb_int_fire=Entry(int_fire_W,textvariable=RESISTANCE_MEMB_INT_FIRE,width=6)
                Vreset_int_fire=Entry(int_fire_W,textvariable=VRESET_INT_FIRE,width=6)
                Vspike_int_fire=Entry(int_fire_W,textvariable=VSPIKE_INT_FIRE,width=6)
                Temps_int_fire=Entry(int_fire_W,textvariable=DUREE_RECORD_INT_FIRE,width=6)
                CstTemps_int_fire=Entry(int_fire_W,textvariable=CONSTANTETEMPS_INT_FIRE,width=6)

                Int_fire_veuillez_choisir_resolution_graph.place(x=0,y=500)
                Resolution_graph_int_fire.place(x=110, y=500)

                Int_fire_veuillez_choisir_durée_record.place(x=0,y=550)
                Temps_int_fire.place(x=110,y=550)

                Int_fire_veuillez_choisir_durée_stim.place(x=0,y=700)
                Durée_stim_int_fire.place(x=110,y=700)

                Int_fire_veuillez_choisir_debut_stim.place(x=0,y=600)
                Start_int_fire.place(x=110,y=600)

                Int_fire_veuillez_choisir_I_input.place(x=0,y=650)
                I_input_int_fire.place(x=110,y=650)

                Int_fire_veuillez_choisir_VThreshold.place(x=400,y=500)
                Vthreshold_int_fire.place(x=510,y=500)

                Int_fire_veuillez_choisir_cst_temps.place(x=400,y=700)
                CstTemps_int_fire.place(x=510,y=700)

                Int_fire_veuillez_choisir_Vrepos.place(x=400,y=550)
                Vrepos_int_fire.place(x=510,y=550)

                Int_fire_veuillez_choisir_R_memb.place(x=400,y=600)
                Resistance_memb_int_fire.place(x=510,y=600)

                Int_fire_veuillez_choisir_Vreset.place(x=400,y=650)
                Vreset_int_fire.place(x=510,y=650)

                Int_fire_veuillez_choisir_Vspike.place(x=0,y=750)
                Vspike_int_fire.place(x=110,y=750)

                fig_int_fire = Figure(figsize=(5, 4), dpi=100)
                canvas_int_fire = FigureCanvasTkAgg(fig_int_fire, master=int_fire_W)              
                canvas_int_fire.get_tk_widget().pack(side=TOP, fill=BOTH, expand=False)
                toolbar = NavigationToolbar2Tk(canvas_int_fire, int_fire_W, pack_toolbar=False)
                toolbar.update()
                toolbar.pack(side=TOP, fill=X)



                def Neuron_integrate_fire():
                   
                    DT_int_fire=float(Resolution_graph_int_fire.get())
                    Temps_record_int_fire=float(Temps_int_fire.get())
                    Stim_duration=float(Durée_stim_int_fire.get())
                    dt_int_fire=int((Temps_record_int_fire*1000/(DT_int_fire/10))+1)
                    time_int_fire=np.linspace(0,Temps_record_int_fire*1000,dt_int_fire)
                    # Vecteur Time crée avec le temps d'integration prit en compte 

                    start_stim_int_fire=int(Start_int_fire.get())
                    end_stim_fire=int(Stim_duration+start_stim_int_fire)
                    I_Value_int_fire=int(I_input_int_fire.get())


                    def valeur(time_int_fire=time_int_fire,start_stim_int_fire=start_stim_int_fire, end_stim_fire=end_stim_fire ,I_Value_int_fire=I_Value_int_fire):

                        I_int_fire=[]
                        for i in range (len(time_int_fire)):
                            if start_stim_int_fire< time_int_fire[i]< end_stim_fire:
                                I_int_fire.append(I_Value_int_fire)
                            else:
                                I_int_fire.append(0)

                        return I_int_fire
                    I_int_fire_leaky=valeur()
                    def leaky_IF(DT_int_fire=float(Resolution_graph_int_fire.get()),time_int_fire=time_int_fire, I_int_fire_leaky=I_int_fire_leaky, tau=float(CstTemps_int_fire.get()), R=float(Resistance_memb_int_fire.get()), Vthreshold=float(Vthreshold_int_fire.get()), Vreset=float(Vreset_int_fire.get()), Vspike=float(Vspike_int_fire.get()), VRest=float(Vrepos_int_fire.get())):
                        V_int_fire = np.ones_like(time_int_fire)*VRest

                        for a in range(len(time_int_fire)-1):
                            dV_int_fire = (DT_int_fire/10)*((-(V_int_fire[a] - VRest) + R*I_int_fire_leaky[a])/tau)
                            V_int_fire[a+1] = V_int_fire[a] + dV_int_fire

                            if V_int_fire[a]>Vthreshold:
                                V_int_fire[a+1] = Vspike
                            if V_int_fire[a] == Vspike:
                                V_int_fire[a+1] = Vreset
                        return V_int_fire

                    ax_int_fire_leaky_IF=fig_int_fire.add_subplot(211)
                    ax_int_fire_leaky_IF.plot(time_int_fire,leaky_IF())
                    ax_int_fire_leaky_IF.set_xlabel("Time [ms]")
                    ax_int_fire_leaky_IF.set_ylabel("Potentiel")

                    ax_int_fire_istim = fig_int_fire.add_subplot(212)
                    ax_int_fire_istim.plot(time_int_fire,valeur())
                    ax_int_fire_istim.set_xlabel("Time [ms]")
                    ax_int_fire_istim.set_ylabel("Intensité")
                    toolbar.update()
                    toolbar.pack(side=TOP, fill=X)
                    canvas_int_fire.draw()

                    return 
                Int_fire_Actualiser=Button(int_fire_W,text="Calculer",command=Neuron_integrate_fire)
                Int_fire_Actualiser.place(x=350,y=750)
                
            def Window_model_iz():
                IZ_W=Toplevel(root,bg="light blue")
                IZ_W.title("Modele de Neurone: \n Izhikevich")
                IZ_W.geometry("800x780");IZ_W.resizable(False, False);IZ_W.geometry("+400+400")
                
                
                
                RESOLUTION_GRAPH_IZ=IntVar()
                START_IZ=IntVar()
                DUREESTIM_IZ=IntVar()
                I_INPUT_IZ=IntVar()
                DUREE_RECORD_IZ=IntVar()
                A_IZ=IntVar()
                B_IZ=IntVar()
                C_IZ=IntVar()
                D_IZ=IntVar()
                
                IZ_veuillez_choisir_resolution_graph=Label(IZ_W,text="Resol graph(ms)")
                IZ_veuillez_choisir_debut_stim=Label(IZ_W,text="Début stim (ms)")
                IZ_veuillez_choisir_durée_stim=Label(IZ_W,text="Durée Stim (ms)")
                IZ_veuillez_choisir_I_input=Label(IZ_W,text="Intensité stim(pA)")
                IZ_veuillez_choisir_durée_record=Label(IZ_W,text='Temps record (s)')
                IZ_veuillez_choisir_A=Label(IZ_W,text="Valeur variable A")
                IZ_veuillez_choisir_B=Label(IZ_W,text="Valeur variable B")
                IZ_veuillez_choisir_C=Label(IZ_W,text="Valeur variable C")
                IZ_veuillez_choisir_D=Label(IZ_W,text="Valeur variable D")

                Resolution_graph_iz=Entry(IZ_W,textvariable=RESOLUTION_GRAPH_IZ,width=6)
                Start_iz=Entry(IZ_W,textvariable=START_IZ,width=6)
                Durée_stim_iz=Entry(IZ_W,textvariable=DUREESTIM_IZ,width=6)
                I_input_iz=Entry(IZ_W,textvariable=I_INPUT_IZ,width=6)
                Temps_iz=Entry(IZ_W,textvariable=DUREE_RECORD_IZ,width=6)
                A_iz=Entry(IZ_W,textvariable=A_IZ,width=6)
                B_iz=Entry(IZ_W,textvariable=B_IZ,width=6)
                C_iz=Entry(IZ_W,textvariable=C_IZ,width=6)
                D_iz=Entry(IZ_W,textvariable=D_IZ,width=6)
                
                IZ_veuillez_choisir_resolution_graph.place(x=0,y=500)
                Resolution_graph_iz.place(x=110, y=500)

                IZ_veuillez_choisir_durée_record.place(x=0,y=550)
                Temps_iz.place(x=110,y=550)

                IZ_veuillez_choisir_durée_stim.place(x=0,y=700)
                Durée_stim_iz.place(x=110,y=700)

                IZ_veuillez_choisir_debut_stim.place(x=0,y=600)
                Start_iz.place(x=110,y=600)

                IZ_veuillez_choisir_I_input.place(x=0,y=650)
                I_input_iz.place(x=110,y=650)
                
                IZ_veuillez_choisir_A.place(x=400,y=500)
                A_iz.place(x=510,y=500)
                
                IZ_veuillez_choisir_B.place(x=400,y=550)
                B_iz.place(x=510,y=550)
                
                IZ_veuillez_choisir_C.place(x=400,y=600)
                C_iz.place(x=510,y=600)

                IZ_veuillez_choisir_D.place(x=400,y=650)
                D_iz.place(x=510,y=650)
                
                fig_iz = Figure(figsize=(5, 4), dpi=100)
                canvas_iz = FigureCanvasTkAgg(fig_iz, master=IZ_W)              
                canvas_iz.get_tk_widget().pack(side=TOP, fill=BOTH, expand=False)
                toolbar = NavigationToolbar2Tk(canvas_iz, IZ_W, pack_toolbar=False)
                toolbar.update()
                toolbar.pack(side=TOP, fill=X)
                
                def Neuron_IZ():
                   
                    DT_iz=float(Resolution_graph_iz.get())
                    Temps_record_iz=float(Temps_iz.get())
                    Stim_duration=float(Durée_stim_iz.get())
                    dt_iz=int((Temps_record_iz*1000/(DT_iz/10))+1)
                    time_iz=np.linspace(0,Temps_record_iz*1000,dt_iz)
                    # Vecteur Time crée avec le temps d'integration prit en compte 

                    start_stim_iz=int(Start_iz.get())
                    end_stim_iz=int(Stim_duration+start_stim_iz)
                    I_Value_iz=int(I_input_iz.get())


                    def valeur_iz(time_iz=time_iz,start_stim_iz=start_stim_iz, end_stim_iz=end_stim_iz ,I_Value_iz=I_Value_iz):

                        I_iz=[]
                        for i in range (len(time_iz)):
                            if start_stim_iz< time_iz[i]< end_stim_iz:
                                I_iz.append(I_Value_iz)
                            else:
                                I_iz.append(0)

                        return I_iz
                    
                    par_iz=(float(A_iz.get()),float(B_iz.get()),float(C_iz.get()),float(D_iz.get()))
                    
                    def Iz(par_iz=par_iz,v0_iz=-65., Vspike_iz=30., DT_iz=DT_iz,Inp_iz=valeur_iz()):
                        v_iz = [v0_iz]
                        u_iz = [v_iz[0] * 0.2]

                        a,b,c,d = par_iz
                        def vdot_iz(v,u,I):
                            return 0.04*v*v + 5.*v + 140. - u + I
                        def udot_iz(v,u):
                            return a * (b*v - u)
                        def reset_iz(v,u):
                            return (c, u+d)
                        for t in range (len(time_iz)-1):
                            if (v_iz[t] >= Vspike_iz):
                                v_iz_new, u_iz_new = reset_iz(v_iz[t], u_iz[t])
                                v_iz[t] = Vspike_iz
                            else:
                                v_iz_new = v_iz[t] + (DT_iz/10) * vdot_iz(v_iz[t],u_iz[t],Inp_iz[t])
                                u_iz_new = u_iz[t] + (DT_iz/10) * udot_iz(v_iz[t], u_iz[t])

                            v_iz.append(v_iz_new)
                            u_iz.append(u_iz_new)
                        return v_iz,u_iz
                    
                    V_IZ,U_IZ=Iz()
                    ax_IZ_iz=fig_iz.add_subplot(211)
                    ax_IZ_iz.plot(time_iz,V_IZ)
                    ax_IZ_iz.set_xlabel("Time [ms]")
                    ax_IZ_iz.set_ylabel("Potentiel")

                    ax_IZ_istim = fig_iz.add_subplot(212)
                    ax_IZ_istim.plot(time_iz,valeur_iz())
                    ax_IZ_istim.set_xlabel("Time [ms]")
                    ax_IZ_istim.set_ylabel("Intensité")
                    toolbar.update()
                    toolbar.pack(side=TOP, fill=X)
                    canvas_iz.draw()
                    return
                



                
                
                IZ_Actualiser=Button(IZ_W,text="Calculer",command=Neuron_IZ)
                IZ_Actualiser.place(x=350,y=750)
            
            dev=Button(root1,image=photo,command=show1)
            dev.place(x=250, y=300)    
            Fermer_app=Button(root1,text="Fermer",command=root.destroy );Fermer_app.pack(side=BOTTOM)
            bouton_mod_int_fire=Button(root1,text="Modèle integrated & fire",bg='#81c6eb',command=Window_model_int_and_fire)
            bouton_mod_int_fire.pack(padx=200,pady=100)
            bouton_mod_iz=Button(root1,text="Modèle Izhikevich",bg='#81c6eb',command=Window_model_iz)
            bouton_mod_iz.place(x=320,y=250)
        demarrer_app()
           
Soumettre_mdp=Button(root,text="Soumettre",command=verifmdp)
Soumettre_mdp.pack(side=BOTTOM)
root.mainloop()
```
